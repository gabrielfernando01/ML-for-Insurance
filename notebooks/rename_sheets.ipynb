```python
import zipfile
import shutil
from pathlib import Path

RAW    = Path("../data/raw")
RENAME = {
    "Emision":    "emision",
    "Siniestros": "siniestros",
    "Comisiones": "comisiones",
}

for xlsx_file in RAW.glob("*.xlsx"):
    tmp = xlsx_file.with_suffix(".tmp.xlsx")
    
    with zipfile.ZipFile(xlsx_file, "r") as zin, \
         zipfile.ZipFile(tmp, "w", zipfile.ZIP_DEFLATED) as zout:
        
        for item in zin.infolist():
            data = zin.read(item.filename)
            
            # Renombrar en xl/workbook.xml (donde viven los sheet names)
            if item.filename == "xl/workbook.xml":
                for old, new in RENAME.items():
                    data = data.replace(
                        f'name="{old}"'.encode(),
                        f'name="{new}"'.encode()
                    )
            
            zout.writestr(item, data)
    
    xlsx_file.unlink()
    tmp.rename(xlsx_file)
    print(f"✓ {xlsx_file.name}")

print("Listo.")
```
