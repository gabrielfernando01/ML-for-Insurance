## Pipeline RAW - Resume 

> Tenemos cuatro `db's` en excel abput `siniestraliad` from `2021` to `2024`. Empezamos extraer cada hoja con un `for` on dierectory `../data/row/*` 

```python
import pandas as pd
from pathlib import Path

RAW       = Path("../data/raw")
PROCESSED = Path("../data/processed")
SHEETS    = ["emision", "siniestros", "comisiones"]
```

```python
# Extraccion xlsxs a dataframes
frames = {sheet: [] for sheet in SHEETS}

for path in sorted(RAW.glob("*.xlsx")):
    anio = int(''.join(filter(str.isdigit, path.stem))[:4])
    print(f"\n{anio} | {path.name}")
    for sheet in SHEETS:
        df = pd.read_excel(path, sheet_name=sheet)  # header=0 por default
        df.insert(0, "ANIO", anio)
        frames[sheet].append(df)
        print(f"  {sheet}: {df.shape}")
```

```python
# Concatenate frames and save to CSV
for sheet in SHEETS:
    df_total = pd.concat(frames[sheet], ignore_index=True)
    out = PROCESSED / f"{sheet.lower()}_total.csv"
    df_total.to_csv(out, index=False)
    print(f"✓ {out.name}  {df_total.shape}")
```

```python
# Load csv files into dataframes
dfs = {}

for sheet in SHEETS:
    dfs[sheet] = pd.read_csv(PROCESSED / f"{sheet}_total.csv", dtype=str)
    print(f"✓ {sheet}: {dfs[sheet].shape}")
```
